### Problem 1

Use the Laplace transform to verify that the step response of the system $\dot{y} + y = x$ is $y(t) = (1 - e^{-t})u(t)$. Be sure to indicate the regions of convergence of any functions in the s-plane.

The Laplacian of $x$ and the transfer function $H$ are as follows:

#### \begin{align}  X = sY + Y\\ H = \frac{1}{s+1}  \end{align}

The step response can be attained as follows:

\begin{align}
u * h (t) &= \mathscr{L}^{-1} \Big\{ \frac{1}{s} H\Big\}\\
          &= \mathscr{L}^{-1} \Big\{ \frac{1}{s} \cdot \frac{1}{s+1} \Big\}\\
          &= \mathscr{L}^{-1} \Big\{ \frac{A_1}{s} + \frac{A_2}{s+1} \Big\}
\end{align}

The terms $A_1$ and $A_2$ can be determined via partial fraction expansions.

\begin{align}
\frac{1}{s + 1} = A_1 + \frac{sA_1}{s + 1} &\rightarrow s = 0, \: A_1 = 1\\
\frac{1}{s} = \frac{A_1(s+1)}{s} + A_2 &\rightarrow s = -1, \: A_2 = -1
\end{align}

Going back to the expression for the Laplacion from before...

\begin{align}
\mathscr{L}^{-1} \Big\{ \frac{A_1}{s} + \frac{A_2}{s+1} \Big\} &= \mathscr{L}^{-1} \Big\{ \frac{1}{s} - \frac{1}{s+1} \Big\}\\
&= u(t) - e^{-t}u(t)\\
&= (1 - e^{-t})u(t)
\end{align}

The step response attained from this method matches with the one in the question.

### Problem 2

a. Find the DC gain of the system $Y(s) / Y_{sp}(s)$ below if you use an integral controller $K(s) = K_i/s$ for any $H(s)$. Does it depend on the value of $K_i$?

\begin{align}
\frac{Y(s)}{Y_{sp}(s)} = \frac{KH}{1 + KH} = \frac{\frac{K_i}{s}H(s)}{1 + \frac{K_i}{s}H(s)} = \frac{K_iH(s)}{s + K_iH(s)}
\end{align}

The DC gain of the system is the limit of the transfer function as $s$ goes to zero.

\begin{align}
\lim_{x\to0} \frac{Y(s)}{Y_{sp}(s)} = \frac{K_iH(s)}{0+K_iH(s)} = 1
\end{align}

The DC gain does not depend on the value of $K_i$.

b. Assume $H(s) = \frac{1/\tau}{s + 1/\tau}$. Find $Y(s)/Y_{sp}(s)$. Find the pole(s) of the system assuming $K \gg 1/\tau$. Draw a pole-zero diagram in this case.

\begin{align}
\frac{Y(s)}{Y_{sp}(s)} = \frac{\frac{1/\tau}{s + 1/\tau}K_i}{s + \frac{1/\tau}{s + 1/\tau}K_i} = \frac{K_i/\tau}{s^2 + s/\tau + K_i/\tau}
\end{align}

The poles can be attained by setting the denominator to zero:

###### \begin{align} s = \Bigg(-1/\tau \pm \sqrt{1/\tau^2 - 4K_i/\tau}\Bigg)\: /\: 2 \approx \Bigg(-1/\tau \pm \sqrt{-4K_i/\tau}\Bigg)\: / \: 2 = -\frac{1}{2\tau} \pm j\sqrt{\frac{K_i}{\tau}} \end{align}

Since $K \gg 1/\tau$, the $1/\tau^2$ term under the square root is extremely small and can effectively be ignored.

There are also no zeros in this system, since the numerator of the transfer function does not depend on $s$.

The pole-zero diagram consists of two poles at the following coordinates:

$(-1/2\tau, \sqrt{K_i/\tau})$ and $(-1/2\tau, -\sqrt{K_i/\tau})$

The region of convergence is $Re\{s\} > -1/2\tau$

### a. $\frac{s}{s+1}$

```python
sys = signal.lti([1, 0], [1, 1])
combinedplot(sys)
```

1st order numerator -> 1 zero, 1st order denominator -> 1 pole

This system behaves like a high pass filter, since it removes all of the higher frequencies of the system. The Fourier transform also exists because the imaginary axis lies within the region of convergence: $Re\{s\} > -1$. The step response decays to zero after jumping up to $1$ and is pretty stable.

### b. $\frac{s}{s^2+100s+1}$

```python
sys = signal.lti([1, 0], [1, 100, 1])
combinedplot(sys)
```

1st order numerator -> 1 zero, 2nd order denominator -> 2 poles

This system behaves like a bandstop filter, as it only preserves the intermediate frequencies, as shown in the Bode plot. Although it appears that one of the poles is on $(0, \: 0)$, its x-coordinate is actually $(-100 \pm \sqrt{100^2 - 4})\:/\:2 \approx -0.01$. The region of convergence is therefore $Re\{s\} > -0.01$, and since the imaginary axis lies in the ROC, the Fourier transform exists. The step response is also stable and decays to 0.

### c. $\frac{s}{s^2+s+1}$

```python
sys = signal.lti([1, 0], [1, 1, 1])
combinedplot(sys)
```

1st order numerator -> 1 zero, 2nd order denominator -> 2 poles

This system also behaves like a bandstop filter and cuts off frequencies that are far away from $1\: rad/s$. In the pole-zero diagram, the poles are clearly in the left-hand plane, and the region of convergence is $Re\{s\} > -0.5$, so the Fourier transform exists. The step-response is also stable and converges towards 0.

### d. $\frac{s}{s^2+0.1s+1}$

The step response is also stable, since it converges to 0, but it oscillates a lot in the beginning.

b. Use the pole-zero map to show the effect of using proportional control on this system. Show the step response of at least two feedback gains to illustrate. Can you stabilize the system?

The new transfer function becomes the following:

### $\frac{Y}{Y_{sp}} = \frac{KH}{1+KH} = \frac{K_p / (s^2-0.01s+1)}{1 + K_p / (s^2-0.01s+1)} = \frac{K_p}{s^2-0.01s+1+K_p}$

The poles for this system are:

#### $s = (0.01 \pm \sqrt{0.0001 - 4(1+K_p)})\: /\: 2 = 0.005 \pm \frac{1}{2} \sqrt{-3.9999 - 4K_p}$

```python
# Proportionality Constant = 10
K_p = 10
sys = signal.lti([K_p], [1, -0.01, 1+K_p])
t = np.linspace(0, 100, 500)

_t, s = signal.step(sys, T=t)
timeplot(t, s, label='step response')
```

```python
pzmap(sys)
```

```python
# Proportionality Constant = 40

K_p = 40
sys = signal.lti([K_p], [1, -0.01, 1+K_p])
t = np.linspace(0, 100, 500)

_t, s = signal.step(sys, T=t)
timeplot(t, s, label='step response')
```

```python
pzmap(sys)
```

By increasing the proportionality constant, the frequency of the signal gets increased. It does nothing to change the fact that the system diverges over time. This makes sense, since the real component of the poles does not change as $K_p$ changes, so it is impossible to shift the region of convergence.

c. Repeat part B using integral control.

The new transfer function is the following:

### $\frac{Y}{Y_{sp}} = \frac{KH}{1 + KH} = \frac{\frac{K_i}{s(s^2 - 0.01s + 1)}}{1 + \frac{K_i}{s(s^2 - 0.01s + 1)}} = \frac{K_i}{s^3 - 0.01s^2 + s + K_i}$

```python
# Integral Constant = 0.1
K_i = 0.1
sys = signal.lti([K_i], [1, -0.01, 1, K_i])
t = np.linspace(0, 100, 500)

_t, s = signal.step(sys, T=t)
timeplot(t, s, label='step response')
```

The integral control changes the spacing between the real values of the positive and negative poles. Thus, the Fourier transform does not exist, regardless of the value of $K_i$, and the system will not stabilize.

d. Repeat part B using differential control.

The transfer function becomes:

### $\frac{Y}{Y_{sp}} = \frac{KH}{1+KH} = \frac{K_ds / (s^2-0.01s+1)}{1 + K_ds / (s^2-0.01s+1)} = \frac{K_ds}{s^2+(K_d-0.01)s+1}$

The poles are:

#### $s = (-K_d + 0.01 \pm \sqrt{(K_d - 0.01)^2 - 4})\: / \: 2$

$K_d$ must exceed 0.01 in order for the real component to become less than zero so that the poles exist in the left-hand plane

```python
# Differential Constant = 1
K_d = 1
sys = signal.lti([K_d, 0], [1, -0.01 + K_d, 1])
t = np.linspace(0, 100, 500)

_t, s = signal.step(sys, T=t)
timeplot(t, s, label='step response')
```

```python
pzmap(sys)
```

```python
# Differential Constant = 0.001
K_d = 0.001
sys = signal.lti([K_d, 0], [1, -0.01 + K_d, 1])
t = np.linspace(0, 100, 500)

_t, s = signal.step(sys, T=t)
timeplot(t, s, label='step response')
```

```python
pzmap(sys)
```

As expected, when $K_d$ is less than 0.01, the system fails to stabilize because the real component is in the right-hand plane. However, when $K_d$ is large enough to push the region of convergence into the negative x-axis region, the system stabilizes and the step response goes to zero.

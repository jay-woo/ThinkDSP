1) Following the example in chap07.ipynb, make a synthesis matrix (the IDFT matrix) using a different range of frequencies, and see what difference it makes. The usual version of IDFT uses frequencies 0 through N-1.  What happens if you go from N to 2N-1?

```python
import thinkdsp, thinkplot
import numpy, scipy.signal

PI2 = 2 * numpy.pi

# Generates synthesis matrix that maps to frequencies in the range 0 through N-1

N = 4
ts = numpy.arange(N) / float(N)
freqs = numpy.arange(N)
args = numpy.outer(ts, freqs)
M = numpy.exp(1j * PI2 * args)
M
```

```python
# Generates synthesis matrix that maps to frequencies in the range N through 2N-1

N = 4
ts = numpy.arange(N) / float(N)
freqs = numpy.arange(N, 2*N)
args = numpy.outer(ts, freqs)
M2 = numpy.exp(1j * PI2 * args)
M2
```

```python
# Checks that the two different synthesis matrices are approximately equal

def approx_equal(a, b, tol=1e-10):
    return sum(abs(a-b)) < tol

approx_equal(M, M2)
```

Even after changing the range of frequencies of the waves being synthesized together, the synthesis matrix does not change at all. This is due to the fact that its just the phases that are changing, rather than the amplitudes. Only the exponent of the expression $Ae^{2\pi i ft}$ changes, as $f$ changes. The amplitude $A$, however, does not change.

2) In thinkdsp.py, the Spectrum methods low_pass and high_pass are ideal filters, in some sense: in the passband they are perfectly flat, and in the stopband they attenuate to 0 (or whatever value you want). Invoking the principle that for every filter there is a corresponding window, see if you can figure out what window corresponds to these ideal filters.

```python
# Taken from chap07.ipynb
def synthesize2(amps, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(1j * PI2 * args)
    ys = numpy.dot(M, amps)
    return ys

ts = numpy.linspace(0, 25, 11250)

# Generates a lowpass window
amps_lowpass = numpy.concatenate((numpy.ones(50), numpy.zeros(50)))
ys_lowpass = synthesize2(amps_lowpass, numpy.linspace(0,1,100), ts)

# Generates a highpass window
amps_highpass = numpy.concatenate((numpy.zeros(50), numpy.ones(50)))
ys_highpass = synthesize2(amps_highpass, numpy.linspace(0,1,100), ts)

thinkplot.plot(ts, ys_lowpass)
thinkplot.config(xlim=[0,10],
                 xlabel='Time (s)',
                 title='Lowpass Window')
```

```python
thinkplot.plot(ts, ys_highpass)
thinkplot.config(xlim=[0,10],
                 xlabel='Time (s)',
                 title='Highpass Window')
```

The lowpass window generated above appears to be the function $\frac{\sin{x}}{x}$. The highpass window, on the other hand, appears to be a more complicated blend of sines and cosines, while also decaying at a rate proportional to $1/x$. 

3) So far we have focused attention on the effect filters have on the magnitude of each frequency component, but they also affect the phase offset.  For the boxcar and Gaussian windows, plot the phase shift as a function of frequency.

```python
# Generates a padded boxcar window
boxcar = numpy.ones(10)
boxcar /= sum(boxcar)
boxcar = numpy.concatenate((boxcar, numpy.zeros(200)))

# Generates a padded Gaussian window
gaussian = scipy.signal.gaussian(M=11, std=2)
gaussian /= sum(gaussian)
gaussian = numpy.concatenate((gaussian, numpy.zeros(200)))
```

```python
# Plots the spectrum of the boxcar window
boxcar_wave = thinkdsp.Wave(boxcar, framerate=44100)
boxcar_spectrum = boxcar_wave.make_spectrum()
boxcar_spectrum.plot()
thinkplot.config(xlabel='Frequency (Hz)',
                 ylabel='Amplitude',
                 title='Spectrum of the Boxcar Window')
```

```python
# Plots the phase offset vs. the frequencies for the boxcar window
boxcar_phases = numpy.angle(boxcar_spectrum.hs)
thinkplot.plot(numpy.linspace(0, 44100/2, len(boxcar_phases)), boxcar_phases)
thinkplot.config(xlabel='Frequency (Hz)',
                 ylabel='Phase Offset (rad)',
                 title='Phase Offset of the Boxcar Window')
```

```python
# Plots the spectrum of the Gaussian window
gaussian_wave = thinkdsp.Wave(gaussian, framerate=44100)
gaussian_spectrum = gaussian_wave.make_spectrum()
gaussian_spectrum.plot()
thinkplot.config(xlabel='Frequency (Hz)',
                 ylabel='Amplitude',
                 title='Spectrum of the Gaussian Window')
```

```python
# Plots the phase offset of the Gaussian window
gaussian_phases = numpy.angle(gaussian_spectrum.hs)
thinkplot.plot(gaussian_phases)
thinkplot.config(xlabel='Frequency (Hz)',
                 ylabel='Phase Offset (rad)',
                 title='Phase Offset of the Gaussian Window')
```

```python
# Reinitializes the boxcar and Gaussian windows
boxcar = numpy.ones(11)
boxcar /= sum(boxcar)

gaussian = scipy.signal.gaussian(M=11, std=2)
gaussian /= sum(gaussian)

# Generates a square wave and calculates the phase offsets for each ferquency value
wave = thinkdsp.SquareSignal(freq=440).make_wave(framerate=44100).segment(duration=0.01)
wave_phases = numpy.angle(numpy.fft.fft(wave.ys))
thinkplot.plot(numpy.linspace(0, wave.framerate/2, 441), wave_phases)
thinkplot.config(xlim=[0,10000],
                 xlabel='Frequency (Hz)',
                 ylabel='Phase Offset (rad)',
                 title='Phase Offset of the Square Wave')
```

```python
# Calculates the phase offsets after convolving with the boxcar window
boxcar_wave = thinkdsp.Wave(numpy.convolve(wave.ys, boxcar, mode='valid'), framerate=44100)
boxcar_phases = numpy.angle(numpy.fft.fft(boxcar_wave.ys))
thinkplot.plot(numpy.linspace(0, wave.framerate/2, len(boxcar_phases)), boxcar_phases)
thinkplot.config(xlim=[0,10000],
                 xlabel='Frequency (Hz)',
                 ylabel='Phase Offset (rad)',
                 title='Phase Offset of the Signal w/ Boxcar Window')
```

```python
# Calculates the phase offsets after convolving with the Gaussian window
gaussian_wave = thinkdsp.Wave(numpy.convolve(wave.ys, gaussian, mode='valid'), framerate=44100)
gaussian_phases = numpy.angle(gaussian_wave.make_spectrum().hs)
thinkplot.plot(numpy.linspace(0, wave.framerate/2, len(gaussian_phases)), gaussian_phases)
thinkplot.config(xlim=[0,20000],
                 xlabel='Frequency (Hz)',
                 ylabel='Phase Offset (rad)',
                 title='Phase Offset of the Signal w/ Gaussian Window')
```

After applying the boxcar and Gaussian windows to the square wave, the phase offsets do not change at all. The general shape is the same across all of the plots for the phase offset.

4) Taking the difference of successive elements in a wave array has the same effect as convolution with the window [1, -1].  The filter for this window is an increasing function of f, so you can think of diff as a high-pass filter. One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f.

If we use a better approximation of a derivative, we might expect the corresponding filter to be a better approximation of a line.  The window [-1/2, 2, -3/2] is a second order forward difference approximation of the first derivative.  Compute its DFT and see what the filter looks like.

Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window?

```python
first_deriv1 = numpy.concatenate(([1, -1], numpy.zeros(500)))
first_deriv2 = numpy.concatenate(([-0.5, 2, -1.5], numpy.zeros(500)))
second_deriv = numpy.concatenate(([1.0, -2.0, 1.0], numpy.zeros(500)))

first_deriv_wave1 = thinkdsp.Wave(first_deriv1, framerate=44100)
first_deriv_wave2 = thinkdsp.Wave(first_deriv2, framerate=44100)
second_deriv_wave = thinkdsp.Wave(second_deriv, framerate=44100)
```

```python
# Plots the spectrum of the first derivative window (1st approximation)
first_deriv_wave1.make_spectrum().plot()
thinkplot.config(xlabel='Frequency (Hz)',
                 ylabel='Amplitude',
                 title='Spectrum of the First Derivative Window (v.1)')
```

```python
# Plots the spectrum of the first derivative window (2nd approximation)
first_deriv_wave2.make_spectrum().plot()
thinkplot.config(xlabel='Frequency (Hz)',
                 ylabel='Amplitude',
                 title='Spectrum of the First Derivative Window (v.2)')
```

```python
# Plots the spectrum of the second derivative window
second_deriv_wave.make_spectrum().plot()
thinkplot.config(xlabel='Frequency (Hz)',
                 ylabel='Amplitude',
                 title='Spectrum of the Second Derivative Window')
```

As expected, the filters that correspond to the first and second derivatives look like high-pass filters. All of the low frequencies get attenuated, while the high frequencies stay about the same. The second approximation is more accurate than the first one, since the low frequencies get attenuated more.

The second derivative filter looks the way it does because it is essentially a high-pass filter over a high-pass filter. The higher frequencies get attenuated less and less, while the lower frequencies get attenuated even more. As a result, there is more of a curvature to the spectrum.

5) The gunshot recording I used in chap09.ipynb was clipped, which means that the loudest part gets cut off, which makes the rest of the sound seem louder than it should be, which exaggerates the reverb of the space. But it looks like there is a better version of this recording at http://www.freesound.org/people/Kleeb/sounds/180960/. Use this recording to replace the one I used and see if it tones down the reverb a little. Also, apply the same effect to the saxophone recording.

```python
# Plots the impulse response of the new gunshot sound
response = thinkdsp.read_wave('180960__kleeb__gunshot.wav')
response = response.segment(start=0.14)
response.normalize()
response.plot()
thinkplot.config(xlim=[0.0, 2.0],
                 ylim=[-1.05, 1.05])
```

```python
response.make_audio()
```

```python
# Plots the spectrum of the impulse response
transfer = response.make_spectrum()
transfer.plot()
```

```python
# Splits the saxophone wave into chunks so the filter can be applied to each chunk
sax = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav').segment(start=0.25)
num_segments = len(sax.ys) / len(response.ys)
sax.ys = sax.ys[:num_segments*len(response.ys)]
sax_vals = numpy.split(sax.ys, len(response.ys))
sax.make_audio()
```

```python
# Applies the filter to each chunk of the audio file
for i in range(num_segments):
    sax_segment = thinkdsp.Wave(sax.ys[i * len(response.ys) : (i+1) * len(response.ys)]-1, framerate=response.framerate)
    spectrum = sax_segment.make_spectrum()
    
    output = (spectrum * transfer).make_wave()
    output.normalize()
    
    if i == 0:
        final = output
    else:
        final = final | output

final.make_audio()
```

6) Use spectrum.differentiate() to compute the derivative of a ParabolicSignal.  Hint: this works best if the duration of the wave is an integer number of periods.

```python
# Generates 30 periods of a parabolic signal
parabola_sig = thinkdsp.ParabolicSignal(freq=500)
parabola_wav = parabola_sig.make_wave(framerate=40000).segment(start=0, duration=30*parabola_sig.period)
parabola_wav.plot()
thinkplot.config(xlim=[0,0.01],
                 xlabel='Time (s)',
                 title='Parabolic Signal')
```

```python
# Plots the spectrum of the parabolic wave
parabola_spec = parabola_wav.make_spectrum()
parabola_spec.plot()
thinkplot.config(xlim=[0,5000],
                 xlabel='Frequency (Hz)',
                 ylabel='Amplitude',
                 title='Spectrum of the Parabolic Signal')
```

```python
# Calculates the derivative of the parabolic signal by differentiating the spectrum
parabola_spec = parabola_wav.make_spectrum()
parabola_spec.differentiate()
parabola_spec.plot()
thinkplot.config(xlim=[0,5000],
                 xlabel='Frequency (Hz)',
                 ylabel='Amplitude',
                 title='Spectrum of the Differentiated Parabolic Signal')
```

```python
parabola_spec.make_wave().plot()
thinkplot.config(xlim=[0, 0.01],
                 xlabel='Time (s)',
                 title='Differentiated Parabolic Signal')
```

Before taking the derivative of the parabolic signal, the amplitudes fall off at a rate of $\frac{1}{f^2}$. Then, after taking the derivative, the amplitudes appear to fall off at a rate of $\frac{1}{f}$. After converting the spectrum back to the signal, the result is as expected - the derivative of a parabola $x^2$ is proportional to $x$, and since the parabola keeps repeating itself, we should expect to see the same line repeat over and over again.

In both the spectrum and the signal, the behavior of the graph changes after taking the derivative of the parabolic signal, as all of the squared elements end up becoming linear ($\frac{1}{f^2} \rightarrow \frac{1}{f}$ and $mx^2 \rightarrow 2mx$).

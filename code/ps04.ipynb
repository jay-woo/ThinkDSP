1) Please finish this exercise that we started in class: One of the examples in chap07.ipynb compares two mixtures of complex sinusoids, one with offset=0 and one with offset=1.  The result is a different looking waveform, because "rotating" each component by the same phase offset does not shift each component by the same amount of time. Can you construct a vector of complex amplitudes that shifts all components by the same amount of time?

```python
PI2 = 2 * numpy.pi

# Generates a wave of 4 waves with the following attributes:
amps = numpy.array([0.6, 0.25, 0.1, 0.05])
freqs = PI2 * numpy.array([100, 200, 300, 400])

framerate = 11025 * 2
n = framerate / 25
offset = 2

# Synthesizes the original wave
ts = numpy.linspace(0, 1, framerate)
ys = dft.synthesize2(amps, freqs, ts)

# Synthesizes the wave after multiplying the amplitude by e^(i*f*phi)
amps2 = amps * numpy.exp(1j * PI2 * numpy.array(freqs) * offset)
ys2 = dft.synthesize2(amps2, freqs, ts)

thinkplot.plot(ts[:n], ys.real[:n], label='Original Wave')
thinkplot.plot(ts[:n], ys2.real[:n], label='Shifted Wave')
thinkplot.config(legend=True, xlim=[0,0.005],
                 title='Signal Before and After Phase Shift',
                 xlabel='Time (s)', ylabel='y')
```

In order to shift the wave by a certain amount in the time domain, we multiplied the original wave by an exponential function of the folloiwng form: $Ae^{i\phi}$. Since the amplitude is not being changed at all, $A$ is simply 1. The original waveform can be represented as $Be^{i2\pi ft}$. When the two exponentials are multiplied together, the result is $Be^{i2\pi ft + \phi}$. In order for $t$ to be shifted, $\phi$ should be a multiple of $i2\pi f$.

So, $Ae^{i\phi}$ becomes $e^{i2\pi f\phi}$, and the final product becomes $Be^{i2\pi f(t + \phi)}$. This corresponds to a phase shift in the time domain.

2) In Chapter 7 I represented the DFT as matrix multiplication, but the more common presentation is as a summation. Read the Wikipedia page on DFT.  Write an implementation of DFT based on this summation and compare your results to my implementation and/or numpy.fft.fft.

```python
# Calculates the DFT of a wave using the summation method
def dft_new(ys):
    N = len(ys)
    amps = numpy.zeros(N)
    freqs = numpy.arange(N)
    
    for i in range(N):
        amps += ys[i] * numpy.exp(-1j * PI2 * i * numpy.array(freqs) / N)
                
    return amps
```

3) In Chapter 7, I showed how we can express the DFT and inverse DFT as matrix multiplications. These operations are relatively fast, taking time proportional to $N^2$, where $N$ is the length of the wave array. That would be fast enough for many applications, but it turns out that there is a faster algorithm, the Fast Fourier Transform (FFT), which takes time proportional to $N\log{N}$. Read about the FFT at http://en.wikipedia.org/wiki/Cooley-Tukey_FFT_algorithm, and write an implementation. Hint: I suggest you write a simple version as a recursive function; don’t worry about “data reordering, bit reversal, and in-place algorithms”.

```python
def fft(ys):
    N = len(ys)
    
    if N == 1:
        return ys
    else:
        y_even = fft(ys[::2])
        y_odd = fft(ys[1::2])
        
        X = [0] * N
        for k in range(N/2):
            X[k] = y_even[k] + y_odd[k] * numpy.exp(-1j * PI2 * k / N)
            X[k+N/2] = y_even[k] - y_odd[k] * numpy.exp(-1j * PI2 * k / N)
                    
        return X
```

```python
# Clips the ys to be a power of 2 in length and applies a hamming window to reduce leakage
clipped_ys = ys[:2**13] * numpy.hamming(2**13)

# Calculates the FFT and plots the spectrum
amps_fft = fft(clipped_ys)
thinkdsp.Spectrum(amps_fft, framerate*2).plot()
thinkplot.config(xlim=[0,250],
                 xlabel='Frequency (Hz)',
                 ylabel='Amplitude',
                 title='Spectrum of Wave After FFT')
```

The FFT algorithm ran a lot faster than the DFT, due to differences in time complexity. This problem really shows how a faster algorithm can really make a huge difference, because the DFT took much, much longer to compute. The result looks slightly different compared to the result of the DFT - this is probably due to the fact that part of the wave was clipped at the end in order to accomodate for the fact that the FFT works for arrays of length 2 to some exponent.

4) Using the timing code from PS03, compare the performance of my implementation of autocorr from Chapter 5, my fft_autocorr from Chapter 8, and numpy.correlate.  Can you tell whether NumPy is using the FFT algorithm?

```python
# Imports everything that is needed for the 3 tests

import autocorr

def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

def fft_convolve(signal, window):
    fft_signal = numpy.fft.fft(signal)
    fft_window = numpy.fft.fft(window)
    return numpy.fft.ifft(fft_signal * fft_window)

def fft_autocorr(signal):
    N = len(signal)
    window = signal[::-1]
    signal = zero_pad(signal, 2*N)
    window = zero_pad(window, 2*N)

    corrs = fft_convolve(signal, window)
    corrs = corrs[N//2: 3*N//2]
    return corrs
```

```python
# Generates some white noise
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=5.0)

# Generates a list of numbers from 2^4 to 2^12
ns = 2 ** numpy.arange(7, 15)

results1 = []
results2 = []
results3 = []

# Times how long it takes to go through 2^N inputs
for N in ns:
    ys = noise.ys[:N]
    noise_snip = thinkdsp.Wave(ys, noise.framerate)
    
    result1 = %timeit -o autocorr.autocorr(noise_snip)
    result2 = %timeit -o fft_autocorr(noise.ys[:N])
    result3 = %timeit -o numpy.correlate(noise.ys[:N], noise.ys[:N], mode='full')[N/2:]
    
    results1.append(result1)
    results2.append(result2)
    results3.append(result3)

bests1 = [result.best for result in results1]
bests2 = [result.best for result in results2]
bests3 = [result.best for result in results3]

# Plots the computational cost vs. the number of inputs
thinkplot.plot(ns, bests1, label=r'Autocorr')
thinkplot.plot(ns, bests2, label=r'FFT Autocorr')
thinkplot.plot(ns, bests3, label=r'Numpy Corr')

thinkplot.config(xscale='log', yscale='log')
thinkplot.config(xlabel='Number of Inputs',
                 ylabel='Computation Time (s)',
                 title='Spectrum of Original Wave')
```

```python
# Performs a linear fit on each data set

x = numpy.log(ns)
y1 = numpy.log(bests1)
y2 = numpy.log(bests2)
y3 = numpy.log(bests3)

t1 = scipy.stats.linregress(x,y1)
t2 = scipy.stats.linregress(x,y2)
t3 = scipy.stats.linregress(x,y3)

print "SLOPE VALUES"
print "Autocorr: " + str(t1[0])
print "FFT Autocorr: " + str(t2[0])
print "Numpy Corr: " + str(t3[0])
```

It is clear that Numpy does not use the FFT algorithm. The FFT autocorrelation algorithm has a time complexity of about $O(N)$, while the Numpy correlation algorithm has a time complexity of about $O(N^2)$. This is quite a significant difference.

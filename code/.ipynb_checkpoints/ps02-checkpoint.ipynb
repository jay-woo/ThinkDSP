1) Run and listen to the examples in chap03.ipynb.  In the leakage example, try replacing the Hamming window with one of the other windows provided by NumPy, and see what effect they have on leakage.

```python
import thinkdsp
import thinkplot
import thinkstats2
import numpy

# Produces a sine wave that is off by about 1/4 a period

sin_sig = thinkdsp.SinSignal(freq=440)
sin_wav = sin_sig.make_wave(duration=sin_sig.period * 3.5)

sin_wav.plot()
```

```python
# Plots the spectrum of the above sine wave

sin_wav.make_spectrum().plot()
```

```python
# Applies a Bartlett window to the sine wave

sin_wav.window(numpy.bartlett(len(sin_wav.ys)))
sin_wav.make_spectrum().plot()
```

As with the Hamming window, the Bartlett window tapers off the frequencies above or below 440 Hz. The original signal produced a lot of leakage, due to the fact that it was not a continuous, periodic function - there is a discontinuity between the end and the beginning of the signal. The Bartlett window tries to fix this discontinuity by making the beginning and the end of the signal closer to zero, and as a result, the signal looks more like a periodic function and there is less leakage.

2) Write a class called SawtoothChirp that extends Chirp and overrides evaluate to generate a sawtooth waveform with frequency that increases (or decreases) linearly. Plot the spectrogram of this signal. The effect of aliasing should be visible, and if you listen carefully, you can hear it.

```python
# An implementation of a sawtooth chirp signal, which starts with a certain frequency that ramps up linearly over time

class SawtoothChirp(thinkdsp.Chirp):
    def _evaluate(self, ts, freqs):
        dts = numpy.diff(ts)
        dphis =  freqs * dts
        cycles = numpy.cumsum(dphis)
        frac, _ = numpy.modf(cycles)
        ys = self.amp * (2*frac - 1)
        return ys
```

```python
# Plots the beginning portion of the signal, which has a frequency of 400 Hz

sawtooth_chirp = SawtoothChirp(500, 1500)
sawtooth_wave = sawtooth_chirp.make_wave()

sawtooth_wave.segment(0.0, 0.01).plot()
```

```python
# Plots the end portion of the signal, which has a frequency of 800 Hz

sawtooth_wave.segment(0.99, 1).plot()
```

```python
sawtooth_wave.make_audio()
```

```python
# Plots the spectrogram of the sawtooth chirp signal

spectrogram = sawtooth_wave.make_spectrogram(seg_length=512)
spectrogram.plot(high=256)
```

Due to aliasing, there are some audible distortions in the sound of the sawtooth chirp signal. The signal sounds very buzzy, and towards the end, you can clearly hear a second tone in the background. The spectrogram makes the aliasing effect a lot easier to see - as the frequencies get higher, you can see that they "fold back" into lower frequencies when the frequency reaches the framerate limit. The higher frequencies keep reverberating against the limits of the sampling rate.

3)  In musical terminology, a “glissando” is a note that slides from one pitch to another, so it is similar to a chirp. A trombone player can play a glissando by extending the trombone slide while blowing continuously. As the slide extends, the total length of the tube gets longer, and the resulting pitch is inversely proportional to length. Assuming that the player moves the slide at a constant speed, how does frequency vary with time? Is a trombone glissando more like a linear or exponential chirp? Write a function that simulates a trombone glissando from C3 up to F3 and back down to C3. C3 is 262 Hz; F3 is 349 Hz.

```python
trombone_wav = thinkdsp.ExpoChirp(start=262, end=349).make_wave() | thinkdsp.ExpoChirp(start=349, end=262).make_wave()
trombone_wav.make_audio()
```

The pitch is inversely proportional to the length of the trombone tube, and since perceived pitch is a logarithmic function of frequency, the frequency should increase exponentially as the pitch increases linearly. Thus, a trombone glissando should sound more like an exponential chirp.

6) In a white noise signal, the mixture of frequencies changes over time. In the long run, we expect the power at all frequencies to be equal, but in any sample, the power at each frequency is random. To estimate the long-term average power at each frequency, we can break a long signal into segments, compute the power spectral density for each segment, and then compute the average across the segments. You can read more about this algorithm at http://en.wikipedia.org/wiki/Bartlett's_method. Apply Bartlett's method to your natural noise source and see if it gives a clearer picture of the relationship between frequency and power.

```python
# This is one of the sounds I used for the previous question
noise = thinkdsp.read_wave('13793__soarer__north-sea.wav').segment(start=0, duration=5.0)

duration = 5.0             # Duration of the sound
N = 10                     # Number of segments to average together
seg_length = duration / N  # Length of each segment

# Performs Bartlett's method
for i in range(N):
    # Takes a short segment of the sound file
    noise_spectrum = noise.segment(start=i * seg_length, duration=seg_length).make_spectrum()
    
    # Squares the amplitudes of each frequency
    noise_power = numpy.square(noise_spectrum.hs)
    noise_framerate = noise_spectrum.framerate
    noise_spectrum = thinkdsp.Spectrum(noise_power, noise_framerate)
    
    # Total_spectrum is used for summing all of the spectra together
    if i == 0:
        total_spectrum = noise_spectrum
    else:
        total_spectrum += noise_spectrum

# Normalizes the total spectrum
total_spectrum = thinkdsp.Spectrum(total_spectrum.hs / N, total_spectrum.framerate)

# Plots the modified spectrum against the original spectrum
total_spectrum.plot(low=1, linewidth=1, alpha=0.5)
noise.make_spectrum().plot_power(low=1, linewidth=1, alpha=0.25)
thinkplot.show(xlabel='Frequency (Hz)',
               ylabel='Power Density',
               xscale='log',
               yscale='log')
```

In the above plot, the lighter color represents the original noise and the darker color represents the modified noise, after using Bartlett's method. The variance in the power density decreased significantly, as evidenced by the fact that the plot doesn't look as thick as the original. The general shape, however, is pretty much the same - if you computed the slope of the modified spectrum, it would probably be similar to that of the original.

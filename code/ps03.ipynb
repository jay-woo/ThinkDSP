2) Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs

bitcoin_lags, bitcoin_corrs = autocorr(wave)

thinkplot.plot(bitcoin_lags, bitcoin_corrs)
thinkplot.show(xlabel='Lag',
               ylabel='Correlation',
               title='Autocorrelation of Bitcoin Data for Various Lag Values')
```

The autocorrelation function looks to be decaying exponentially, and as the lag increases, the correlation flattens out at around 0.1 or 0.2. There appears to be some sort of periodic behavior, as evidenced by the various spikes in the plot. Every so often, the correlation goes up, so there must be certain segments of the bitcoin data that is periodic.

4) Test the algorithmic complexity of analyze1, analyze2 and scipy.fftpack.dct.  You will find code in ps03starter.ipynb that will help with this.

```python
# Test 1: dct.analyze1

# Generates some white noise
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)

# Generates a list of numbers from 2^4 to 2^12
ns = 2 ** numpy.arange(4, 12)
results = []

# Times how long it takes to go through 2^N inputs
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o dct.analyze1(ys, freqs, ts)
    results.append(result)

bests = [result.best for result in results]

# Plots the computational cost vs. the number of inputs
thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log')
```

```python
# Performs a linear fit on the data from test 1

x = numpy.log(ns)
y = numpy.log(bests)
t = scipy.stats.linregress(x,y)

print "Slope: " + str(t[0])
```

```python
# Test 2: dct.analyze2

results = []

for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o dct.analyze2(ys, freqs, ts)
    results.append(result)

bests = [result.best for result in results]

thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log')
```

```python
# Performs a linear fit on the data from test 2

x = numpy.log(ns)
y = numpy.log(bests)
t = scipy.stats.linregress(x,y)

print "Slope: " + str(t[0])
```

```python
# Test 3: scipy.fftpack.dcts

results = []
ns = 2 ** numpy.arange(6, 18)  # I increased the range on this function
                               # When ns is low, the computational speed appears to be relatively constant

for N in ns:
    ys = noise.ys[:N]
    result = %timeit -o scipy.fftpack.dct(ys)
    results.append(result)

bests = [result.best for result in results]

thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log')
```

```python
# Performs a linear fit on the data from test 3
# When n > 10^4, there isn't any more additional data

x = numpy.log(ns[5:8])    # Takes just the middle band, where it appears to be linear
y = numpy.log(bests[5:8])
t = scipy.stats.linregress(x,y)

print "Slope: " + str(t[0])
```

Based on the linear fits, analyze1 < analyze2 < scipy.fftpack.dct, in terms of computational complexity. The fitted lines for analyze1 and analyze2 had slopes of around 2. The dct algorithm from scipy, on the other hand, was much more efficient - the slope of the line was 1, which corresponds to $O(n)$. This is orders of magnitudes faster than the other two algorithms.

5) One of the major applications of the DCT is compression for both sound and images. In its simplest form, DCT-based compression works like this:
1. Break a long signal into segments.
2. Compute the DCT of each segment.
3. Identify frequency components with amplitudes so low they are inaudible, and remove them. Store only the frequencies and amplitudes that remain.
4. To play back the signal, load the frequencies and amplitudes for each segment and apply the inverse DCT.

Implement a version of this algorithm and apply it to a recording of music or speech. How many components can you eliminate before the difference is perceptible?

```python
sax = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav').segment(duration=5.0)

# Compresses a sound file into approximately 1/20th of the original size
def compressSignal(N, wave, percentile):
    segments = []       # List of all of the sound segments
    amplitudes = []     # List of all the amplitudes (1 list per segment)
    frequencies = []    # List of all the frequencies (1 list per segment)
    
    # Divides the wave into N segments and extracts the ampltudes/frequencies
    for i in range(N):
        # thinkdsp.Wave data
        temp = wave.segment(start=wave.duration/N * i, duration=wave.duration/N)
        
        # Amplitude data
        ys = [float(i) for i in temp.ys.tolist()]
        
        # Frequency data
        num_samples = len(ys)
        framerate = temp.framerate
        freq = numpy.linspace(0.0, 0.5*framerate, num_samples)
        
        segments.append(temp)
        amplitudes.append(scipy.fftpack.dct(ys))
        frequencies.append(freq)
        
    # Goes through each list of amplitudes and removes small amplitudes
    for i in range(len(amplitudes)):
        amplitudes[i] = numpy.abs(amplitudes[i])
        
        threshold = numpy.percentile(amplitudes[i], percentile)   # All amplitudes below the 95th percentile are cut out
        deletions = []   # Keeps track of all the deleted amplitudes/frequencies
        
        for j in range(len(amplitudes[i])):
            if amplitudes[i][j] < threshold:
                deletions.append(j)
                
        amplitudes[i] = numpy.delete(amplitudes[i], deletions, None)
        frequencies[i] = numpy.delete(frequencies[i], deletions, None)
                
    return segments, amplitudes, frequencies
```

```python
# Compresses the audio file down, after splitting the file into 10 segments

segments, amplitudes, frequencies = compressSignal(10, sax, 99)

sax.make_spectrum().plot()
thinkplot.show(xlabel='Frequency (Hz)',
               ylabel='Amplitude',
               title='Spectrum of Original Audio',
               xlim=[0, 5000],
               ylim=[0.0, 1.2e8])
```

```python
# Combines all of the audio together into 1 wave and plots the spectrum

for i in range(10):  # Iterates through the 10 segments
    ts = numpy.linspace(0.0, segments[i].duration, sax.framerate/10)
    ys = dct.synthesize1(amplitudes[i], frequencies[i], ts=ts)
    wave = thinkdsp.Wave(ys, framerate=sax.framerate)
    
    if i == 0:
        total = wave
    else:
        total = total | wave
    
total.make_spectrum().plot()
thinkplot.show(xlabel='Frequency (Hz)',
               ylabel='Amplitude',
               title='Spectrum of Compressed Audio')
```

6) Write a class called SawtoothSignal that extends Signal and provides evaluate to evaluate a sawtooth signal.  Note: there is an implementation of SawtoothSignal in thinkdsp.py; you should avoid looking at it until you have written your own.

```python
# Extends the Signal class to produce a sawtooth wave

class SawtoothSignal(thinkdsp.Sinusoid):
    def evaluate(self, ts):
        cycles = self.freq * ts + self.offset / thinkdsp.PI2
        frac, _ = numpy.modf(cycles)
        ys = (frac * 2) - 1
        return ys
```

```python
# Generates a sawtooth signal using the overridden evaluate() function

sawtooth_signal = SawtoothSignal()
sawtooth_signal.plot()
```

Numpy's modf() function returns the fractional portions of the frequencies. These values start at 0 and ramp up to 1, and this pattern repeats for each period. By multiplying the amplitude by 2, and then subtracting 1 from those values, the program produces a sawtooth graph that starts at -1, ramps up to 1, and then repeats that pattern.

7) Compute the spectrum of a sawtooth wave. How does the harmonic structure compare to triangle and square waves?

```python
sawtooth_spectrum = sawtooth_signal.make_wave().make_spectrum()
sawtooth_spectrum.plot()
```

Unlike the triangle and the square waves, the sawtooth wave has both odd and even harmonics of the fundamental frequency, which is 440 Hz in the case of the above spectrum. It also falls off at a rate of $\frac{1}{f}$, which is the same in a square wave.

10)  The triangle and square waves have odd harmonics only; the sawtooth wave has both even and odd harmonics. The harmonics of the square and sawtooth waves drop off in proportion to $\frac{1}{f}$; the harmonics of the triangle wave drop off like $\frac{1}{f^2}$. Can you find a waveform that has even and odd harmonics that drop off like $\frac{1}{f^2}$?

```python
# In this first attempt, I synthesized a series of sine waves,
# that are multiples of a fundamental frequency of 440 Hz.
# The amplitude also drops off at a rate of 1/f^2.

freq = 440
old_freq = freq
new_freq = freq

amp = 1.0
ratio = 1

signals = []

# Produces 15 sine waves that drop off at a rate of 1/f^2
# Harmonic/dominant frequency: 440 Hz
for i in range(15):
    old_freq = new_freq
    new_freq = freq * (i+1)
            
    ratio *= float(new_freq) / old_freq  # Calculates the ratio between two consecutive frequencies
    signals.append(thinkdsp.SinSignal(freq=new_freq, amp=1/ratio**2))

sum_signals = sum(signals)
sum_signals.plot()
```

```python
sum_signals.make_wave().make_spectrum().plot()
```

```python
sum_signals.make_wave().make_audio()
```

```python
# In my second attempt, I multiplied two equal sawtooth signals together.
# Since one sawtooth signal falls off at a rate of 1/f, I tried
# multiplying two together in order to get a fall-off rate of 1/f^2

sawtooth1 = thinkdsp.SawtoothSignal()
sawtooth2 = thinkdsp.SawtoothSignal()

sawtooth_mul = sawtooth1.make_wave() * sawtooth2.make_wave()
sawtooth_mul.make_spectrum().plot()
```
